```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from prophet import Prophet
df = pd.read_csv('FinalWNVData v.1')
```

```python
df2 = pd.DataFrame()
df2['Wnv'] = df.groupby('Date').sum()['WnvPresent']
df2.index = pd.to_datetime(df2.index)
df2['Date'] = df2.index
df3 = pd.DataFrame()
df3['ds'] = df2['Date']
df3['y'] = df2['Wnv']
```

```python
df3.max()
```

```python
prophet = Prophet(daily_seasonality = True,weekly_seasonality=True)
prophet.fit(df3)
```

```python
future = prophet.make_future_dataframe(periods=60,freq='d')
future['cap'] = 38
forecast = prophet.predict(future)
```
